```python
import pandas as pd 
```

```python
df= pd.read_excel('Online Retail.xlsx')
df.head()
```

```python
df.info()
```

```python
df = df.dropna(subset=['CustomerID','Description'])

df= df[~df[ 'InvoiceNo'].astype(str).str.startswith('C')]

df = df[df['Quantity'] > 0]

df = df[df[ 'UnitPrice'] > 0]

df = df.drop_duplicates()

df.info()
```

```python
df
```

```python

df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
total_revenue = df['TotalPrice'].sum()
total_revenue
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['YearMonth'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df[['InvoiceDate','YearMonth']].head()
```

```python
monthly_revenue = df.groupby('YearMonth')['TotalPrice'].sum()
monthly_revenue 
```

```python
import matplotlib.pyplot as plt 

```

```python
monthly_revenue.plot()
plt.title('monthly revenue trend')
plt.xlabel('month')
plt.ylabel('revenue')
plt.show()
```

```python
monthly_revenue = df.groupby('YearMonth')['TotalPrice'].sum()
monthly_revenue.sort_values(ascending=False)
```

```python
top_products_quantity = (
df.groupby('Description')['Quantity'].sum()
    .sort_values(ascending=False)
)
top_products_quantity.head(20)
```

```python
top_products_list= (
df.groupby('Description')['TotalPrice'].sum()
.sort_values(ascending=False)
)
top_products_list.head(20)
```

```python
df['TopProduct'] = df['Description'].apply(lambda x: x if x in top_products_list else 'Other')
df
```

```python
import plotly.express as px


df_filtered = df[(df['Quantity'] > 0) & (df['TotalPrice'] > 0)]

top_revenue_products = df_filtered.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10).index

top_quantity_products = df_filtered.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10).index

def label_product(x):
    if x in top_revenue_products and x in top_quantity_products:
        return 'Top Revenue & Quantity'
    elif x in top_revenue_products:
        return 'Top Revenue'
    elif x in top_quantity_products:
        return 'Top Quantity'
    else:
        return 'Other'

df_filtered['ProductLabel'] = df_filtered['Description'].apply(label_product)

# -----------------------------
# Interactive scatter plot
fig = px.scatter(
    df_filtered,
    x='Quantity',
    y='TotalPrice',
    color='ProductLabel',
    hover_data=['Description', 'Quantity', 'TotalPrice'],
    title='Revenue vs Quantity Sold (Interactive)',
    log_x=True,  # optional: log scale for better visualization
    log_y=True
)
fig.write_html("images/revenue_vs_quantity_interactive.html")
fig.show()
```

```python
import seaborn as sns

df['YearMonthStr'] = df['InvoiceDate'].dt.to_period('M').astype(str)

monthly_quantity = df.groupby(['YearMonthStr', 'Description'])['Quantity'].sum().reset_index()

top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(5).index
monthly_top = monthly_quantity[monthly_quantity['Description'].isin(top_products)]

plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_top, x='YearMonthStr', y='Quantity', hue='Description', marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Quantity Sold for Top Products')
plt.ylabel('Quantity Sold')
plt.xlabel('Month')
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
